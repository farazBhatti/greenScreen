```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Load an image
img = cv2.imread('GreenScreen.jpg')

# Convert the image to HSV color space
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds of the color range you want to threshold
lower_color = np.array([50, 50, 50]) # Replace with your desired values
upper_color = np.array([90, 255, 255]) # Replace with your desired values

# Threshold the image to create a binary mask
mask = cv2.inRange(hsv_img, lower_color, upper_color)

# Apply the mask to the original image
result = cv2.bitwise_and(img, img, mask=mask)

# Display the original image and the result
cv2_imshow(result)
```
